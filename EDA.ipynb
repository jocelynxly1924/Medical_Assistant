```python
import pandas as pd
import json
```

```python
data = []
with open('data/Huatuo26M-Lite/format_data.jsonl','r',encoding='utf-8') as json_file:
    for line in json_file:
        data.append(json.loads(line))
df = pd.DataFrame(data)
```

```python
print('size:', len(df))
labels = df['label'].unique()
print(labels)
```

```python
label_counts = df['label'].value_counts()
print(label_counts)
```

```python
disease_counts = df['related_diseases'].value_counts()
print(disease_counts[:20])
```

```python
target_diseases = ['感冒', '咳嗽', '发烧']
df_respiratory_diseases = df[df['related_diseases'].isin(target_diseases)]
print(df_respiratory_diseases.head(3))
print("data length: ", len(df_respiratory_diseases))
```

```python
with open('data/respiratory_symptoms.jsonl', 'w', encoding='utf-8') as f:
    for _, row in df_respiratory_diseases.iterrows():
        # 只保存需要的字段，按需调整
        item = {
            'question': row['question'],
            'answer': row['answer'],
            'related_diseases': row.get('related_diseases', ''),
            'label': row.get('label', '')
        }
        f.write(json.dumps(item, ensure_ascii=False) + '\n')

# print("已保存到 respiratory_symptoms.jsonl")
```
